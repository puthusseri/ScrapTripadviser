```python
import requests 
from bs4 import BeautifulSoup 
import csv 
import re
from selenium import webdriver
import os
import time
```

```python
listOfUrls = []
listOfUrls.append('https://www.tripadvisor.in/Restaurant_Review-g297637-d6706468-Reviews-or150-Garden_Grille_Restaurant-Thiruvananthapuram_Trivandrum_Thiruvananthapuram_D.html')
listOfUrls.append('https://www.tripadvisor.in/Restaurant_Review-g297637-d7242432-Reviews-or130-Cafe_Jade-Thiruvananthapuram_Trivandrum_Thiruvananthapuram_District_Kerala.html')
listOfUrls.append('https://www.tripadvisor.in/Restaurant_Review-g297637-d3493926-Reviews-or1240-Villa_Maya-Thiruvananthapuram_Trivandrum_Thiruvananthapuram_District_Kerala.html')
```

```python
chromedriver = "./chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
f = open("a.txt","a")
lis = []
for found in listOfUrls:
    while True:
        url=found
        driver.get(url)
        try:
            driver.find_element_by_class_name('ulBlueLinks').click()
            time.sleep(2)
            all_spans = driver.find_elements_by_xpath("//p[@class='partial_entry']")
            for span in all_spans:
                print(span.text)
                st = span.text
                st = st.replace('\n', ' ').replace('\r', '')
                lis.append(st)
                print("\n\n\n\n\n")


            for i in lis:
                f.write(str(i))
                f.write("\n")

            driver.find_element_by_class_name('next').click()
            found=driver.current_url

        except Exception as e:
            print("exit")
            print(e)
            break
    
f.close()

```
